```python
from bs4 import BeautifulSoup as bs
import requests

frstPath = 'https://www.bbc.com/'
basePath = 'https://www.bbc.com/hindi/topics/cvjp25yn1n7t'
#basePath = 'https://www.bbc.com/hindi/topics/cvjp25yn1n7t/page/12'

allLils = [basePath]
```

```python
while True:
    request = requests.get(allLils[-1])
    soup = bs(request.content, 'html5lib')
    try:
        nextLink = soup.select('.gel-layout__item div div div div~ div+ div')
        for i in nextLink:
            link = frstPath + i.select('a')[0].get('href')
            if link not in allLils:
                allLils.append(link)
    except:
        break

print(allLils)
```

```python
news = []
for link in allLils:
    request = requests.get(link)
    soup = bs(request.content, 'html5lib')
    text = 'qa-heading-link lx-stream-post__header-link gs-u-align-middle'

    links = soup.select('a')
    for i in links:
        try:
            if 'qa-heading-link' in i.attrs['class'] or 'lx-stream-post__header-text' in i.attrs['class'] or 'gs-u-align-middle' in i.attrs['class']:
                news.append(frstPath + i.get('href'))
        except:
            pass

print(len(news))
```

```python
allNews = []
tt = ['bbc-1shrd86', 'e1cc2ql70']

for link in news:
    request = requests.get(link)
    soup = bs(request.content, 'html5lib')

    head = soup.select('#content')[0].get_text()

    newss = soup.select('p')
    temp = []
    try:
        for i in newss:
            try:
                if i.attrs['class'] == tt:
                    temp.append(i.get_text())
            except:
                pass
    except:
        pass
    newsss = temp[0]
    allNews.append([head, newsss, 0])
```

```python
import pandas as pd

df = pd.DataFrame(allNews, columns =['Heading', 'News', 'Label'])
df
```

```python
df.to_csv('fake_news.csv', index=False)
```
